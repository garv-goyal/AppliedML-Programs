#### 8. Use the standard deviation method to remove outliers. Apply the method to each feature individually and choose the data points which are an outlier in any of the features to remove. You can choose the lower and upper thresholds. Explain how you decided on the thresholds and what were the outlier samples removed from the dataset. (Don't forget the labels and patient IDs!)


```python
feature_columns = df.columns[2:] 

outliers_index = pd.Series(False, index=df.index)

threshold = 3

for col in feature_columns:
    feature_data = df[col]
    data_mean, data_std = feature_data.mean(), feature_data.std()
    lower_bound = data_mean - threshold * data_std
    upper_bound = data_mean + threshold * data_std

    # for datasets that lie outside lower bound and upper bound
    is_outlier = (feature_data < lower_bound) | (feature_data > upper_bound)
    
    outliers_index = outliers_index | is_outlier

    print(f"Number of outliers in feature '{col}': {is_outlier.sum()}\n")

# to get indices of all samples
outlier_sample_indices = df.index[outliers_index].tolist()

print("Total number of outlier samples to remove:", len(outlier_sample_indices))
print("\nIndices of outlier samples:", outlier_sample_indices)

df = df.drop(index=outlier_sample_indices).reset_index(drop=True)

print("\nDataframe shape after removing outliers:", df.shape)
```

**I set the threshold at 3, as used in statistical analyses which I think can identify values that are different from mean**

#### 9. Select a subset of the features using the ANOVA test. Plot a histogram of all the ANOVA scores and decide which features to keep. Print which features those are. 

```python
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.preprocessing import LabelEncoder

X = df.iloc[:, 2:]  
y = df.iloc[:, 1]  

label_encoder = LabelEncoder()
y_encode = label_encoder.fit_transform(y)  

fs = SelectKBest(score_func=f_classif, k='all')
fs.fit(X, y_encode)

scores = fs.scores_    # get the f-scores for each feature 

plt.bar(range(len(scores)), scores)
plt.show()

feature_scores = pd.Series(scores, index=X.columns)
print("\nANOVA F-scores for each feature:")
print(feature_scores)

selected_features = feature_scores.index.tolist()    # selects all feature names

df = df.iloc[:, [0, 1] + [df.columns.get_loc(col) for col in selected_features]]
```
